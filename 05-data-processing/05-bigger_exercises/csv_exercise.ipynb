```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
file_path = "suicide-rate-by-country-2024.csv"
data = pd.read_csv(file_path, index_col=0)
data.head()
```

```python
suicide_rates = data["SuicideRate_BothSexes_RatePer100k_2021"].sort_values(ascending=False).head(20)
plt.figure(figsize=(12, 6))
plt.bar(suicide_rates.index, suicide_rates.values, color="skyblue")

plt.title("Suicide Rates by Country (2021)")
plt.xlabel("Country")
plt.ylabel("Suicide Rate per 100k")
plt.xticks(rotation=45)

plt.show()
```

```python
countries = ["Belgium", "United States", "Japan", "Germany"]
years = ["2019", "2020", "2021"]
plt.figure(figsize=(10, 6))

for country in countries:
    if country in data.index:
        rates = [data[f"SuicideRate_BothSexes_RatePer100k_{year}"][country] for year in years]
        plt.plot(years, rates, marker=".", label=country)

plt.title("Suicide Rate Trends (2019-2021)")
plt.xlabel("Year")
plt.ylabel("Suicide Rate per 100k")
plt.legend()
plt.grid(True)
plt.legend(loc="upper left", bbox_to_anchor=(1, 1))

plt.show()
```
